```python
Continent_geometry = unary_union(ContShape.geometry)
EEZshapefile["continental_geometry"] = EEZshapefile.geometry.intersection(Continent_geometry)
EEZshapefile["EEZ_geometry"] = EEZshapefile.geometry
EEZshapefile["marine_geometry"] = EEZshapefile.geometry.difference(EEZshapefile.continental_geometry)
EEZshapefile.drop(columns="geometry", axis=1, inplace=True)
EEZshapefile.set_geometry("marine_geometry", inplace=True)
```

Ahora agrego los datos de comutacion entre celdas a eezs. Para eso traduzco las celdas a un int de eez y las agrego

```python
df_eezs = df.copy()
df_eezs["O05"] = df["O05"].apply(lambda x: cells_to_eez[x])
df_eezs["D05"] = df["D05"].apply(lambda x: cells_to_eez[x])

# aggregate data that share Taxon, T, O05, D05 and sum the ndispl column
df_eezs_agg = df_eezs.groupby(["Taxon", "T", "O05", "D05"]).agg({'ndispl':"sum"}).reset_index()
display(df_eezs_agg)


```

```python

# get the unique combinations of Taxon, T and O05 on the aggregated data
taxon_t_o05 = df_eezs_agg[["Taxon", "T", "O05"]].drop_duplicates()
# for each combinantion, compute the sum of ndispl for all the D05
# and normalize the ndispl column by the sum of ndispl for all the D05

df_eezs_agg_norm = df_eezs_agg.copy()
df_eezs_agg_norm["ndispl"] = df_eezs_agg_norm["ndispl"].astype(float)

for idx, row in taxon_t_o05.iterrows():
    columns_idx = (df_eezs_agg["Taxon"] == row["Taxon"]) & (df_eezs_agg["T"] == row["T"]) & (df_eezs_agg["O05"] == row["O05"])
    df_subset = df_eezs_agg[columns_idx].copy()
    sum_ndispl = df_subset["ndispl"].sum()
    df_eezs_agg_norm.loc[columns_idx, "ndispl"] = df_eezs_agg_norm.loc[columns_idx, "ndispl"] / sum_ndispl

display(df_eezs_agg_norm)

assert df_eezs_agg_norm["ndispl"].sum() == len(taxon_t_o05)
```

```python
RepPoints = EEZshapefile.loc[:, ["Country", "continental_geometry"]].copy().set_geometry("continental_geometry", crs="epsg:4326")
RepPoints.geometry = RepPoints.geometry.representative_point() 
geodetic_lines = []
for i in range(len(RepPoints)):
    if RepPoints.geometry.iloc[i].is_empty:
        continue
    for j in range(i + 1, len(RepPoints)):
        if RepPoints.geometry.iloc[j].is_empty:
            continue
        source_country = RepPoints['Country'].iloc[i]
        target_country = RepPoints['Country'].iloc[j]
        source_geometry = RepPoints.geometry.iloc[i]
        target_geometry = RepPoints.geometry.iloc[j]
        geodetic_line = LineString([source_geometry, target_geometry])
        geodetic_lines.append([source_country + ";" + target_country, geodetic_line])

columns = ['Source Country;Target Country', 'geometry']
geodetic_gdf = gpd.GeoDataFrame(pd.DataFrame(geodetic_lines, columns=columns), geometry='geometry')
geodetic_gdf.crs = RepPoints.crs  # Set the same coordinate reference system as the original GeoDataFrame.
geodetic_gdf
```

```python
list_taxon = df_eezs_agg_norm["Taxon"].unique()
list_t = df_eezs_agg_norm["T"].unique()

taxon = "Cetaceans"
t= 1

for taxon in list_taxon:
    print(taxon)
    for t in list_t:
        print(t)
        
        G = nx.DiGraph()
        sources = df_eezs_agg_norm[(df_eezs_agg_norm["Taxon"] == taxon) & (df_eezs_agg_norm["T"] == t)]["O05"].unique()
        targets = df_eezs_agg_norm[(df_eezs_agg_norm["Taxon"] == taxon) & (df_eezs_agg_norm["T"] == t)]["D05"].unique()
        nodes = np.unique(np.concatenate([sources, targets]))

        G.add_nodes_from(nodes)

        for idx, row in df_eezs_agg_norm[(df_eezs_agg_norm["Taxon"] == taxon) & (df_eezs_agg_norm["T"] == t)].iterrows():
            G.add_edge(row["O05"], row["D05"], weight=row["ndispl"])

        # remove self loops
        G.remove_edges_from(nx.selfloop_edges(G))
        # remove node "-1"
        G.remove_node("-1")
        # remove nodes that are not in pos_node
        G.remove_nodes_from([node for node in G.nodes() if node not in node_pos])


        fig, ax = plt.subplots(figsize=(10,10), dpi=100, facecolor='white')
        ax.set_facecolor('black')       
        # For WG84
        ax.set_xlim(-180, 180)
        ax.set_ylim(-90, 90)

        # Plot marine part of the EEZ
        EEZshapefile.set_geometry("marine_geometry", inplace=True, crs="epsg:4326")
        # EEZshapefile.to_crs("esri:54009").plot(aspect=1, edgecolor="#666666", facecolor="#444444", linewidth=0.1, ax=ax)
        EEZshapefile.plot(
            aspect=1,
            edgecolor="white",
            facecolor="none",
            linewidth=0.1,
            ax=ax)
        
        # Plot continental part of the EEZ
        EEZshapefile.set_geometry("continental_geometry", inplace=True, crs="epsg:4326")
        # EEZshapefile.to_crs("esri:54009").plot(aspect=1, edgecolor="#666666", facecolor="white", linewidth=0.1, ax=ax)
        EEZshapefile.plot(
            aspect=1,
            edgecolor="white",
            facecolor="none",
            linewidth=0.1,
            ax=ax
            )

        nx.draw_networkx_edges(G,
                               node_size=0,
                               pos=node_pos,
                               edge_color="red",
                               width=[G[u][v]['weight']*3 for u,v in G.edges()],
                               arrowstyle="->",
                               connectionstyle='arc3,rad=0.2',
                               ax=ax,
                               min_source_margin=0,
                               min_target_margin=0,)
        # do not show the axis labels and ticks
        ax.set_xticks([])
        ax.set_yticks([])



        fig.savefig(f"images/{taxon}_{t}_days.png", dpi=300, bbox_inches="tight")
        fig.savefig(f"images/{taxon}_{t}_days.pdf", dpi=300, bbox_inches="tight")
```
